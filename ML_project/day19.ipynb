```python
cat_col = df.select_dtypes(include="O")
cat_col.head()
```

```python
num_col = df.select_dtypes(exclude="O")
num_col.head()
```

```python
cat_col.drop(['bike_name','city'],axis =1 )
```

```python
unique_owner = cat_col['owner'].unique()
unique_owner
```

```python
n = 1
dt = {}
for i in unique_owner:
    dt[i] = n
    n+=1
```

```python
dt
```

```python
cat_col['owner'] = cat_col['owner'].map(dt)
```

```python
unique_brands = cat_col['brand'].unique()
unique_brands
```

```python
n = 1
dt2 = {}
for i in unique_brands:
    dt2[i] = n
    n+=1
# {brand:i for i , brand in enumerate(unique_brands)}
# list comprehention
```

```python
dt2
```

```python
cat_col['brand'] =  cat_col['brand'].map(dt2)
```

```python
cat_col = cat_col.drop(['bike_name','city'],axis =1 )
```

```python
cat_col
```

```python
pd.get_dummies(df).astype(int)
```

```python
df2 = pd.concat([cat_col,num_col],axis =1)
```

```python
df2.head(2)
```

```python
df2.isnull().sum()
```

```python
# x , y independent , dependent
x = df2.drop('price', axis = 'columns')
y = df2[['price']]
```

```python
#train testing split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test =  train_test_split(x,y,test_size=0.2) #80% train , 20 % testing
```

```python
x_train
```

### MODEL TRAINNING

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train) #trainning
```

### MODEL EVALUATION

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
# model score should be above 90%
# model underfit or overfit
# train & testing score differenece should not be more than 5%
```

```python
#model overfit
    # trainning score -->.high
    # test score --> low
    # (overfit condition)
#model underfit 
    #trainning score >> low
    # testing score>>> low/high
```
