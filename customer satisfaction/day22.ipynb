```python
df['Gender'] = df['Gender'].map({'Female':0,'Male':1})
df
```

```python
df3 =pd.get_dummies(df,drop_first="if_binary").astype(int)
df3
```

```python
df3.isnull().sum()
```

```python
x= df3.drop('satisfaction_satisfied',axis = 1)
y = df3[['satisfaction_satisfied']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg =LogisticRegression()
```

```python
lg.fit(x_train,y_train)
```

```python
print('training score :',round(lg.score(x_train,y_train),2)*100)
print('testing score :',round(lg.score(x_test,y_test),2)*100)
```

```python
#get prediction
prediction = lg.predict(x_test)
#addign prediction column with actual data set
y_test['satisfaction_prediction'] = prediction
y_test.head(25)
```

```python
from  sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test['satisfaction_satisfied'], y_test['satisfaction_prediction'])
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test['satisfaction_satisfied'], y_test['satisfaction_prediction'])
```
